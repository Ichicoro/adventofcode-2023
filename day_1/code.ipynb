### Part 1

```python
import re

with open("input.txt", "r") as f:
    total = 0
    for line in f:
        results = re.findall(pattern=r'(\d)', string=line)
        string = results[0] + results[-1]
        print(results, string)
        total += int(string)
    print(total)
```

### Part 2

```python
import regex as re

with open("input.txt", "r") as f:
    total = 0
    for line in f:
        results = re.findall(pattern=r'(\d|one|two|three|four|five|six|seven|eight|nine)', string=line, overlapped=True)
        for i, result in enumerate(results):
            try:
                int(result)
            except ValueError:
                results[i] = str(['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine'].index(result) + 1)
        string = results[0] + results[-1]
        print(line, results, string)
        total += int(string)
    print(total)
```
