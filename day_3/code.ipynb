### Part 1

```python
import re

with open("input.txt", "r") as file:
    valid_parts = {}
    prev_line = ""
    row = 1
    curr_line = file.readline().rstrip()
    while curr_line:
        next_line = file.readline().rstrip()
        print(f"prev_line {row-1} = {prev_line}")
        print(f"curr_line {row} = {curr_line}")
        print(f"next_line {row+1} = {next_line}")

        parts_re = re.compile(pattern=r"(\d+)")
        parts = []
        for i, line in enumerate([prev_line, curr_line, next_line]):
            if line and len(line) > 0:
                for match in parts_re.finditer(string=line):
                    parts.append((row + i - 1, match))

        mark_re = re.compile(pattern=r"([^\w\d\s.])")
        mark_locations = []
        for match in mark_re.finditer(string=curr_line):
            mark_locations.append(match.start(0))

        if len(mark_locations) == 0:
            print(f"\tno marks found")
            prev_line = curr_line
            curr_line = next_line
            row += 1
            print("============================================================")
            continue

        for mark in mark_locations:
            print(f"\tdoing mark {mark} ({curr_line[mark]})")
            for part_row, part in parts:
                part_num = part.group(0)
                (start, end) = part.span()
                print(f"\t\tmark: {mark} ({curr_line[mark]}), part: {part.group(0)}, start: {start}, end: {end}")
                if mark >= (start - 1) and mark <= (end):
                    print(f"\t\t\tadd to engine {part_num}")
                    valid_parts[f"{part_row}-{start}"] = int(part_num)

        prev_line = curr_line
        curr_line = next_line
        row += 1
        print("============================================================")

    print(f"valid parts: {valid_parts}")
    print(f"valid parts sum: {sum([int(part) for part in valid_parts.values()])}")
```

### Part 2

```python
import re

with open("input.txt", "r") as file:
    total_mul = 0
    prev_line = ""
    row = 1
    curr_line = file.readline().rstrip()
    while curr_line:
        next_line = file.readline().rstrip()
        print(f"prev_line {row-1} = {prev_line}")
        print(f"curr_line {row} = {curr_line}")
        print(f"next_line {row+1} = {next_line}")

        parts_re = re.compile(pattern=r"(\d+)")
        parts = []
        for i, line in enumerate([prev_line, curr_line, next_line]):
            if line and len(line) > 0:
                for match in parts_re.finditer(string=line):
                    parts.append((row + i - 1, match))

        mark_re = re.compile(pattern=r"([*])")
        mark_locations = []
        for match in mark_re.finditer(string=curr_line):
            mark_locations.append(match.start(0))

        if len(mark_locations) == 0:
            print(f"\tno marks found")
            prev_line = curr_line
            curr_line = next_line
            row += 1
            print("============================================================")
            continue

        for mark in mark_locations:
            print(f"\tdoing mark {mark} ({curr_line[mark]})")
            valid_parts = {}
            for part_row, part in parts:
                part_num = part.group(0)
                start, end = part.span()
                print(f"\t\tmark: {mark} ({curr_line[mark]}), part: {part.group(0)}, start: {start}, end: {end}")
                if mark >= (start - 1) and mark <= (end):
                    print(f"\t\t\tmatched_with {part_num}")
                    valid_parts[f"{part_row}-{start}"] = int(part_num)
            if len(valid_parts) > 1:
                 mul = 1
                 for part in valid_parts.values():
                     mul *= part
                 total_mul += mul

        prev_line = curr_line
        curr_line = next_line
        row += 1
        print("============================================================")

    print(f"total mul: {total_mul}")
```
